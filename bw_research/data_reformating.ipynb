## Save cleaned data to csvs/measurements/raw

```python
mf = MyFrame('csvs/measurements_clean.csv')
sessions = mf.getSessions()
```

```python
sessions[1].data
```

```python
# Save each session in different csv file.
# Name of each file will be first 6 letters of signature + id of session
save_path = 'csvs/measurements/raw/'
relay_dic = {}
count = 0

for s in sessions:
    cur_relay = s.relay
    cur_data = s.data
    session_id = relay_dic.get(cur_relay, 0)
    relay_dic.update({cur_relay : session_id+1})
    
    df_save = cur_data[['datetime', 'relay', 'rtt']]
    df_save.to_csv(save_path + cur_relay[:6] + '_' + str(session_id) + '.csv', index=False)
    count +=1
    
print(count)
```

## Filter data and save to filtered folder

```python
# Get names of csv files to be filtered
file_names = [x for x in os.listdir(save_path) if '.csv' in x]
save_path_filtered = 'csvs/measurements/filtered/'

for f in file_names:
    df = pd.read_csv(save_path+f)
    df['rtt'] = getFilteredSignal(df['rtt'])
    df.to_csv(save_path_filtered + f, index=False)
    
```
